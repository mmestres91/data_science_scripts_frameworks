```python
data_base = data_pull
data_base_legacy = data_legacy
```

```python
#Feature Engineering - Create Fabric Indicator from Services Used Column

word_list = ['Fabric']
data_base['fabric_ind'] = data_base.services_used.str.extract('({})'.format('|'.join(word_list)), 
                         expand=False).str.lower().fillna('')
data_base['fabric_ind'] = data_base['fabric_ind'].replace('fabric', 'Yes')
data_base['fabric_ind'] = data_base['fabric_ind'].replace('', 'No')

#Feature Engineering - Create Rental Indicator 
word_list2 = ['Enterprise Holdings']
data_base['rental_ind'] = data_base.services_used.str.extract('({})'.format('|'.join(word_list2)),
                        expand=False).str.lower().fillna('')
data_base['rental_ind'] = data_base['rental_ind'].replace('Enterprise Holdings', 'Yes')
data_base['rental_ind'] = data_base['rental_ind'].replace('enterprise Holdings', 'Yes')
data_base['rental_ind'] = data_base['rental_ind'].replace('', 'No')
```

```python
#Data Cleaning - Legacy

data_base_legacy = data_base_legacy.drop_duplicates(subset=['claim_number'], keep='first')
data_base_legacy = data_base_legacy.drop(columns=['claim_number', 'claim_number_no_dash'])
data_base_legacy = data_base_legacy.rename(columns={'modified_ces': 'ces'})
data_base_legacy = data_base_legacy.rename(columns={'claim_number_1': 'claim_number'})
data_base_legacy = data_base_legacy.rename(columns={'method_of_inspection': 'auto_method_of_inspection'})
data_base_legacy
```

```python
#Concatenate USP & Legacy Surveys

data_base = data_base.drop_duplicates(subset=['claim_number'], keep='first')
data_base = data_base.loc[data_base['line_of_business'].isin(['Personal Auto Line'])]
data_base = pd.concat([data_base, data_base_legacy], axis=0, ignore_index=True)
data_base
```

```python
# Amount of Variation
# Data Cleaning/Manipulation
# Feature Engineering - Age & Customer Tenure

data = data_base.drop_duplicates(subset=['claim_number'], keep='first')
data = data.loc[~(data['ces'].isna())]
def tenure(length):
    return 2022 - length
data['original_effective_datetime'] = pd.to_datetime(data['original_effective_datetime'], infer_datetime_format=True)
data['original_effective_datetime_year'] = data['original_effective_datetime'].dt.year
data['original_effective_datetime_year'] = data['original_effective_datetime_year'].astype('Int64')
data['customer_tenure'] = data['original_effective_datetime_year'].apply(tenure)
data = data.drop(columns=['original_effective_datetime'])
data = data.drop(columns=['original_effective_datetime_year'])
data['customer_tenure'] = data['customer_tenure'].astype('Int64')
def age(length):
    return 2022 - length
data['dateofbirth'] = data['dateofbirth'].astype('string')
data["dateofbirth"] = data["dateofbirth"].str.replace('00:00:00', '', regex=False)
data = data.loc[~(data['dateofbirth'].str.contains('1-01-01 00:00:00', case=False, regex=False, na=False))]
data = data.loc[~(data['dateofbirth'].isin(['0001-01-01 ']))]
data['dateofbirth'] = pd.to_datetime(data['dateofbirth'], infer_datetime_format=True)
data['dob_year'] = data['dateofbirth'].dt.year
data['age'] = data['dob_year'].apply(age)
data = data.drop(columns=['dateofbirth', 'dob_year'])
data['age'] = data['age'].astype('Int64')
data = data.replace(0E-9, np.nan)
def laborday(hours):
    return hours / 4
data['labor_hrs'] = pd.to_numeric(data['labor_hrs'], downcast='float', errors='coerce')
data['labor_days'] = data['labor_hrs'].apply(laborday)
sum_column = data["days_fnol_to_first_close"] + data["labor_days"]
data["auto_claim_ct"] = sum_column
data = data.drop(columns=['fnol'])
data['paid_loss_amount'] = pd.to_numeric(data['paid_loss_amount'], downcast='float', errors='coerce')
data['reserve_loss_amount'] = pd.to_numeric(data['reserve_loss_amount'], downcast='float', errors='coerce')
data = data.drop(columns=['tl_valuation_request_return_days'])
for column_name in ['insured_fault_pct', 'deductible_amt', 'supp_cnt', 'repair_days', 'file_start_to_complete_days', 'loss_to_reported_days', 'loss_to_estimate_start_days', 'est_start_to_file_complete_days', 'loss_reported_to_assignment_days', 'assignment_sent_to_retrieved_days', 'estimate_start_to_estimate_complete_days', 'first_supplement_to_last_supplement_days', 'repair_cost_to_regional_value_pct', 'regional_value_amt']:
    data[column_name] = data[column_name].astype('float64')
data = data.drop(columns=['services_used'])
for column_name in ['paid_loss_amount', 'reserve_loss_amount', 'labor_hrs']:
    data[column_name] = data[column_name].astype('float64')
data = data.drop(columns=['auto_claim_ct'])
data = data.drop(columns=['labor_hrs'])
data.paid_loss_amount = data.paid_loss_amount.round(2)
data.labor_days = data.labor_days.round(2) 
data = data.drop(columns=['line_of_business'])
data = data.drop(columns=['vehicle_age'])
data["benchmark_column"] = pd.NaT
data[['benchmark_column']] = data[['benchmark_column']].fillna('x')
data = data.loc[data['auto_method_of_inspection'].isin(['Desk', 'Field', 'DRP', 'Open Shop', 'Customer Virtual Photos'])]
data = data.drop(columns=['emergency_transport_used'])
data
```
